## Metrics for evaluating a binary classification model

* __Accuracy__ 
* Precision
* Recall
* F1-Score

__Precision-Recall__ is a useful measure of success of prediction when the classes are very imbalanced.

### Confusion Matrix

||Predicted: Yes |Predicted: No|
|--|
|__Actual: Yes__| True Positive| False Negative|
|__Actual: No__| False Positive| True Negative|


---

Precision = How many selected items are relevant (_how many people from those that we said would survive actually survived_)

Precision = TP / (TP + FP)


---

Recall = How many relevant items are selected (_how many people from those that actually survived we said would survive_)

Recall = TP / (TP + FN)

F-Score = Combines _Precision_ and _Recall_ into a single value

F-Score = 2 \* (Precision \* Recall) / (Precision + Recall)

## Metrics for evaluating a binary classification model

Consider the metric that interests us the most. Do we care more about false positives or about false negatives? Also consider other ways of evaluating performance such as ROC and Lift.

Data from [MSDN - How to evaluate model performance in Azure Machine Learning](https://docs.microsoft.com/en-us/azure/machine-learning/studio/evaluate-model-performance)
* Accuracy
 * Accuracy is simply the proportion of correctly classified instances. It is usually the first metric you look at when evaluating a classifier. However, when the test data is unbalanced (where most of the instances belong to one of the classes), or you are more interested in the performance on either one of the classes, accuracy doesn't really capture the effectiveness of a classifier. In the income level classification scenario, assume you are testing on some data where 99% of the instances represent people who earn less than or equal to 50K per year. It is possible to achieve a 0.99 accuracy by predicting the class <=50K for all instances. The classifier in this case appears to be doing a good job overall, but in reality, it fails to classify any of the high-income individuals (the 1%) correctly.
* Precision
 * A very natural question is: Out of the individuals whom the model predicted to be earning >50K (TP+FP), how many were classified correctly (TP)? This question can be answered by looking at the Precision of the model, which is the proportion of positives that are classified correctly: TP/(TP+FP). 
* Recall
 * Another common question is Out of all the high earning employees with income >50k (TP+FN), how many did the classifier classify correctly (TP). This is actually the Recall, or the true positive rate: TP/(TP+FN) of the classifier. You might notice that there is an obvious trade-off between precision and recall. For example, given a relatively balanced dataset, a classifier that predicts mostly positive instances, would have a high recall, but a rather low precision as many of the negative instances would be misclassified resulting in a large number of false positives.
* F-Score
 * Another related metric that is often used is the F1 Score, which takes both precision and recall into consideration. It is the harmonic mean of these 2 metrics and is computed as such: F1 = 2 (precision x recall) / (precision + recall). The F1 score is a good way to summarize the evaluation in a single number, but it's always a good practice to look at both precision and recall together to better understand how a classifier behaves.
* AUC
 * In addition, one can inspect the true positive rate vs. the false positive rate in the Receiver Operating Characteristic (ROC) curve and the corresponding Area Under the Curve (AUC) value. The closer this curve is to the upper left corner, the better the classifier's performance is (that is maximizing the true positive rate while minimizing the false positive rate). Curves that are close to the diagonal of the plot, result from classifiers that tend to make predictions that are close to random guessing.

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

newplot = lambda width=10,height=5: plt.figure(figsize=(width,height)).gca() 
```

## Extract Deck from Cabin

```python
df['Deck'] = df['Cabin'].str.extract('(^[A-Z])', expand = True)

df['Deck'].describe()
```

```python
df['Deck'].value_counts()
```

```python
df['Deck'].value_counts().plot.bar(ax = newplot())
```

```python
df['Deck'].fillna('X', inplace=True)
df['Deck'].value_counts()
```

```python
df[['Fare', 'Deck']].boxplot(by='Deck', ax=newplot())
```

```python
sns.countplot(y='Deck', hue='Survived', data=df)
```

## Analyze Age

```python
g = sns.FacetGrid(df, col='Survived')
g.map(plt.hist, 'Age', bins=20)
```

```python
g = sns.FacetGrid(df, col='Pclass', hue='Survived', palette='Set2')

g.map(plt.hist, 'Age', alpha=.5, bins=20)
g.add_legend();
```

## Apply MICE (Multiple Imputation by Chained Equations) to fill in missing Age values

## Fill in Embark

```python
df[df['Embarked'].isnull()]
```

```python
sns.boxplot(x='Embarked', y='Fare', hue='Sex', data=df[df['Pclass'] == 1], palette='Set1', ax=newplot())
```

```python
df.groupby(['Sex', 'Pclass', 'Embarked'])['Fare'].median()
```

```python
df['Embarked'].fillna('S', inplace=True)
```

## Extract Title from Name

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=True)

pd.crosstab(df['Sex'], df['Title'])
```

```python
df['Title'] = df['Title'].replace('Mlle', 'Miss')
df['Title'] = df['Title'].replace('Ms', 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col', 
                                            'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

pd.crosstab(df['Sex'], df['Title'])

pd.crosstab(df['Survived'], df['Title'])
```

```python
df[['Title', 'Survived']].groupby('Title').mean().plot.bar(ax=newplot())
```

## Combine SibSp and Parch into a single feature

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.tail()
```

```python
df.groupby(['FamilySize'])['Survived'].mean()
```

```python
df.groupby(['FamilySize'], as_index=False)['Survived'].mean()\
    .plot.bar(x='FamilySize', ax=newplot())
```

## Drop unnecessary columns

```python
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch'],\
        axis=1, inplace=True)

df.head()
```

```python

sns.pairplot(df.dropna()[['Survived', 'Age', 'Fare', 'Pclass', 'FamilySize', 'Title', 'Deck']], hue='Survived', plot_kws={'alpha':0.3, 's':80})
```
